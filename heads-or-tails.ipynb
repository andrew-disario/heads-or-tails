```python
import random

player_1 = 'name'
player_1_score = int(0)
player_2 = 'name'
player_2_score = int(0)

number_of_rounds = 3

heads_or_tails = ['HEADS', 'TAILS']

print('Welcome to 2P HEADS OR TAILS (PATENT PENDING!)')

starting_game = True
while starting_game == True:
    player_1 = input('\nEnter Player 1\'s name: ')
    if player_1 != '':
        player_2 = input('Enter Player 2\'s name: ')
        if player_2 != '':
            print(f'\n[New game: {player_1} vs. {player_2}]')
            starting_game = False
        else:
            print('ERROR: Input not recognized. Restarting launch.')
    else:
        print('ERROR: Input not recognized. Restarting launch.')
    
turn = int(2)
while turn != (number_of_rounds + 1) * 2:
    if (turn % 2) == 0:
        print (f'\n[Round {turn//2}, {player_1}\'s turn]')
        guess = input('Enter "0" to guess HEADS or "1" to guess TAILS: ')
        if guess == '0' or guess == '1':
            guess = heads_or_tails[int(guess)]
            result = heads_or_tails[random.randint(0,1)]
            print(f'{player_1}\'s guess: {guess} | Coin flip result: {result}')
            if guess == result:
                print(f'{player_1} was correct and gets one point!')
                player_1_score += 1
            else:
                print(f'{player_1} was not correct and does not get a point.')
            print(f'{player_1}\'s score: {player_1_score} | {player_2}\'s score: {player_2_score}')
            turn += 1            
        else:
            print('ERROR: Input not recognized: Restarting turn.')
    else:
        print (f'\n[Round {turn//2}, {player_2}\'s turn]')
        guess = input('Enter "0" to guess HEADS or "1" to guess TAILS: ')
        if guess == '0' or guess == '1':
            guess = heads_or_tails[int(guess)]
            result = heads_or_tails[random.randint(0,1)]
            print(f'{player_2}\'s guess: {guess} | Coin flip result: {result}')
            if guess == result:
                print(f'{player_2} was correct and gets one point!')
                player_2_score += 1
            else:
                print(f'{player_2} was not correct and does not get a point.')
            print(f'{player_1}\'s score: {player_1_score} | {player_2}\'s score: {player_2_score}')
            turn += 1            
        else:
            print('ERROR: Input not recognized: Restarting turn.')
                
print('\n[Game over]')
print(f'{player_1}\'s final score: {player_1_score} | {player_2}\'s final score: {player_2_score}')
if player_1_score == player_2_score:
    print('It\'s a tie!')
elif player_1_score > player_2_score:
    print(f'{player_1} wins!')
else:
    print(f'{player_2} wins!')
print('\nThanks for playing 2P HEADS OR TAILS (PATENT PENDING!)')
```
